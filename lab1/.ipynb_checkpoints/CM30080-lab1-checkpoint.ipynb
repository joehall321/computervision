### Loading, Writing and Displaying images

We will be using the OpenCV toolkit and matplotlib and numpy libraries in this particular notebook

```python
import matplotlib, cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread('ducks.jpeg')

# show image with matplotlib
plt.imshow(img)
```

```python
# Note that the color order is BGR instead of RGB. Let's fix that to plot it
# color conversion
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# we will mainly be working with grayscale images. Let's do that conversion
# convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# grayscale image represented as a 2-d array
print(gray_img)
```

```python
# have to convert grayscale back to RGB for plt.imshow()
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```

```python
cv2.imwrite('gray-ducks.jpeg',gray_img)
```

```python
# The image array is a regular numpy.ndarray. Hence all the regular numpy operations can be applied
print(type(gray_img))
print(gray_img.shape)
```

```python
# The image is too large. Let's get a small section of the image

small_img = gray_img[1200:1800,1200:1800]
plt.imshow(cv2.cvtColor(small_img, cv2.COLOR_GRAY2RGB))
```

## Filtering Operations

```python
# process with Gaussian blurring, with 7x7 kernel
img_blur = cv2.GaussianBlur(small_img, (7,7), 49)
plt.subplot(121),plt.imshow(cv2.cvtColor(small_img, cv2.COLOR_GRAY2RGB)), plt.title('Original')
plt.subplot(122),plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_GRAY2RGB)), plt.title('Gaussian Blur')
```

```python
cv2.imwrite('blur-ducks.jpeg',img_blur)
```

```python
# Another way to obtain the blur is as follows
kernel = np.ones((7,7),np.float32)/49
box_blur = cv2.filter2D(small_img,-1,kernel)
# This is a box filter with 5x5 box filter
plt.subplot(121),plt.imshow(cv2.cvtColor(small_img, cv2.COLOR_GRAY2RGB)), plt.title('Original')
plt.subplot(122),plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_GRAY2RGB)), plt.title('Box Blur')
```

```python
# As can be achieved through this function

img_blur = cv2.blur(small_img,(7,7))
plt.subplot(121),plt.imshow(cv2.cvtColor(small_img, cv2.COLOR_GRAY2RGB)), plt.title('Original')
plt.subplot(122),plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_GRAY2RGB)), plt.title('Box Blur')
```

## Edge Detection

```python
## Edge Detection is a simple function call

# find binary image with edges
ducks_edges = cv2.Canny(small_img, threshold1=70, threshold2=110)
plt.imshow(cv2.cvtColor(ducks_edges, cv2.COLOR_GRAY2RGB))
```

```python
# Let's compare the effect that blurring has
img_blur = cv2.GaussianBlur(small_img, (7,7), 49)
ducks_edges_Gaussian = cv2.Canny(img_blur, threshold1=70, threshold2=110)
img_blur = cv2.blur(small_img,(7,7))
ducks_edges_box = cv2.Canny(img_blur, threshold1=90, threshold2=110)
plt.subplot(131),plt.imshow(cv2.cvtColor(ducks_edges, cv2.COLOR_GRAY2RGB)), plt.title('Original Edges')
plt.subplot(132),plt.imshow(cv2.cvtColor(ducks_edges_Gaussian, cv2.COLOR_GRAY2RGB)), plt.title('Gaussian Edges')
plt.subplot(133),plt.imshow(cv2.cvtColor(ducks_edges_box, cv2.COLOR_GRAY2RGB)), plt.title('Box Edges')
```

```python
cv2.imwrite('ducks_edges.jpg', ducks_edges)
cv2.imwrite('ducks_edges_Gaussian.jpg', ducks_edges_Gaussian)
cv2.imwrite('ducks_edges_Box.jpg', ducks_edges_box)
```

```python
# Note that we could also obtain the edge detection by using the Sobel Edge Detector. 
#The Sobel edge detector is obtained by using an x gradient and the y gradient. 
# These are obtained as follows
sobelx = cv2.Sobel(small_img, ddepth=cv2.CV_8U, dx=1, dy=0, ksize=3)
sobely = cv2.Sobel(small_img, ddepth=cv2.CV_8U, dx=0, dy=1, ksize=3)
plt.subplot(121),plt.imshow(cv2.cvtColor(sobelx, cv2.COLOR_GRAY2RGB)), plt.title('X Derivatives Sobel')
plt.subplot(122),plt.imshow(cv2.cvtColor(sobely, cv2.COLOR_GRAY2RGB)), plt.title('Y Derivatives Sobel')
```

```python
# As Sobel filter is obtained by using two specific filters, we could also obtain them directly by using the 
# filter parameters directly and convolving them using OpenCV's cv2.filter2D function

kernel_x = np.array([
  [-1, 0, 1],
  [-2, 0, 2],
  [-1, 0, 1]
])
img_x = cv2.filter2D(small_img,-1,kernel_x)
kernel_y = np.array([
  [-1, -2, -1],
  [0, 0, 0],
  [1, 2, 1]
])
img_y = cv2.filter2D(small_img,-1,kernel_y)
plt.subplot(121),plt.imshow(cv2.cvtColor(img_x, cv2.COLOR_GRAY2RGB)), plt.title('X Derivatives Gradient')
plt.subplot(122),plt.imshow(cv2.cvtColor(img_y, cv2.COLOR_GRAY2RGB)), plt.title('Y Derivatives Gradient')
```
